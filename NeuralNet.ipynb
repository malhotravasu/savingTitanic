```python
# Boiler
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers

import tensorflow as tf

df = pd.read_csv('data/temp/train_ready.csv')
rtest = pd.read_csv('data/temp/test_ready.csv')
```

```python
# Features
[df.columns, rtest.columns]
```

```python
# Asserting input
assert(list(df.columns[1:]) == list(rtest.columns[1:]))
```

```python
# Normalizing
scaler = StandardScaler()
scaler.fit(df.iloc[:, 1:])

df.iloc[:, 1:] = scaler.transform(df.iloc[:, 1:])
rtest.iloc[:, 1:] = scaler.transform(rtest.iloc[:, 1:])
```

```python
df.head()
```

```python
rtest.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df['Survived'])
```

```python
model = Sequential()
model.add(Dense(2, input_dim=18, activation='tanh', kernel_regularizer=regularizers.l2(0.0001), activity_regularizer=regularizers.l1(0.0001)))
# model.add(Dense(2, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(df.iloc[:, 1:], df['Survived'], epochs=300, batch_size=64, verbose=1)
# model.fit(X_train, y_train, epochs=300, batch_size=64, verbose=1)
```

```python
predictions = model.predict(X_test).round().astype(int)
print(accuracy_score(y_test, predictions))
```

```python
rtest['Survived'] = model.predict(rtest.iloc[:, 1:]).round().astype(int)
rtest[['PassengerId', 'Survived']].to_csv('data/final/latest.csv', index=False)
```
